# <center><ins>Model Comparison and Selection</ins></center>

--------------------------------------------------------------------------------------------------------

| Model name | Speed (ms) | Training steps | Training loss | Validation loss | mAP @ 0.5IOU | mAP @ 0.75IOU |
| --- | --- | --- | --- | --- | --- | --- |
| Faster R-CNN ResNet101 V1 1024x1024 | 72 | 17000 | 0.167 | 0.185 | 0.9125 | 0.6436 |
| Faster R-CNN Inception ResNet V2 1024x1024 | 236 | 21000 | 0.266 | 0.2731 | 0.90 | 0.606 |


#### <center><b>Table 1</b></center>

### From the table 1, it can be seen that the training and validation losses of <mark>Faster R-CNN ResNet101 V1 1024x1024</mark> object detection model is lesser than the second model in the table. Furthermore, mean average precision of the first model is better than then second model. For a good object detection model mean average precision values should be higher (which shows how well model is going to detect the classes) and losses should be lower. Hence, <mark>Faster R-CNN ResNet101 V1 1024x1024</mark> has been selected as the object detection model for the identification of birds species  as it yields the best results.
